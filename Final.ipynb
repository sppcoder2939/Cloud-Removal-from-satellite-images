```python
import sys
import os
import tarfile
import tifffile as tiff
from PyQt5.QtWidgets import QApplication, QMainWindow, QWidget, QVBoxLayout, QHBoxLayout, QLabel, QPushButton, QFileDialog, QListWidget
from PyQt5.QtCore import Qt
from PyQt5.QtGui import QImage, QPixmap, QPainter
from PyQt5.QtWidgets import QGraphicsScene, QGraphicsView, QGraphicsPixmapItem
import numpy as np
from PIL import Image
import rasterio as rio

class ImageProcessingApp(QMainWindow):
    def __init__(self):
        super(ImageProcessingApp, self).__init__()
        self.initUI()

    def initUI(self):
        self.central_widget = QWidget(self)
        self.setCentralWidget(self.central_widget)

        layout = QVBoxLayout()
        self.central_widget.setLayout(layout)
        
        self.cloudy_folders_paths = []
        self.cloud_free_folders_paths = []
            
        self.selected_cloudy_files_list = QListWidget()
        self.selected_cloud_free_files_list = QListWidget()

        first_row_layout = QHBoxLayout()
        self.cloudy_folders_label = QLabel("Select Cloudy Image Folders (.tar):")
        self.cloud_free_folders_label = QLabel("Select Cloud-Free Image Folders (.tar):")
        
        first_row_layout.addWidget(self.cloudy_folders_label)
        first_row_layout.addWidget(self.cloud_free_folders_label)
        
        third_row_layout = QHBoxLayout()
        third_row_layout.addWidget(self.selected_cloudy_files_list)
        third_row_layout.addWidget(self.selected_cloud_free_files_list)
        
        fourth_row_layout = QHBoxLayout()
        self.cloudy_folders_button = QPushButton("Add Image")
        self.cloudy_folders_button.clicked.connect(self.browse_cloudy_folders)
        self.cloud_free_folders_button = QPushButton("Add Image")
        self.cloud_free_folders_button.clicked.connect(self.browse_cloud_free_folders)
        
        fourth_row_layout.addWidget(self.cloudy_folders_button)
        fourth_row_layout.addWidget(self.cloud_free_folders_button)
        
        # Create and configure the Process button
        self.process_button = QPushButton("Process")
        self.process_button.clicked.connect(self.load_and_display_images)
        
        self.suggestion = QLabel("Instruction: You have to put .tar folder areas simontaneously Example: Put the Cloudy Image and then put the Non Cloud Image\nCaution! Do not put the .tar folder areas in wrong order")
        layout.addWidget(self.suggestion)
        self.space = QLabel("\n")
        layout.addWidget(self.space)
        # Add the "Cloudy" and "Cloud-Free" layouts to the main layout
        layout.addLayout(first_row_layout)
        self.suggestion2 = QLabel("You can dobble click to delete unwanted .tar folder. Please make sure to put .tar folder areas in corrected order")
        layout.addWidget(self.suggestion2)
        layout.addLayout(third_row_layout)
        layout.addLayout(fourth_row_layout)
        
        self.selected_cloudy_files_list.itemDoubleClicked.connect(self.remove_selected_cloudy_file_from_list)
        self.selected_cloud_free_files_list.itemDoubleClicked.connect(self.remove_selected_cloud_free_file_from_list)

        # Add the Process button to the main layout
        layout.addWidget(self.process_button)
        self.preview = QLabel("Preview")
        layout.addWidget(self.preview)

        self.setWindowTitle("Cloud Detection and Removal")
        self.setGeometry(100, 100, 800, 600)

        # Add QGraphicsView for displaying images
        self.graphics_view = CustomGraphicsView(self)  # Use a custom QGraphicsView
        layout.addWidget(self.graphics_view)

        self.scene = QGraphicsScene()
        self.graphics_view.setScene(self.scene)
        self.zoom_factor = 1.0

    def browse_cloudy_folders(self):
        files, _ = QFileDialog.getOpenFileNames(self, "Select Cloudy Image Folders (.tar)", filter="Tar files (*.tar)")
        if files:
            self.cloudy_folders_paths.extend(files)
            # Update the selected files label with the selected cloudy files
            self.update_selected_cloudy_files_list()

    def browse_cloud_free_folders(self):
        files, _ = QFileDialog.getOpenFileNames(self, "Select No CLoud Image Folders (.tar)", filter="Tar files (*.tar)")
        if files:
            self.cloud_free_folders_paths.extend(files)
            # Update the selected files label with the selected cloud-free files
            self.update_selected_cloud_free_files_list()

    def update_selected_cloudy_files_list(self):
        self.selected_cloudy_files_list.clear()
        self.selected_cloudy_files_list.addItems(self.cloudy_folders_paths)

    def update_selected_cloud_free_files_list(self):
        self.selected_cloud_free_files_list.clear()
        self.selected_cloud_free_files_list.addItems(self.cloud_free_folders_paths)
        
    def remove_selected_cloudy_file_from_list(self, item):
        if item.text() in self.cloudy_folders_paths:
            self.cloudy_folders_paths.remove(item.text())
            self.update_selected_cloudy_files_list()

    def remove_selected_cloud_free_file_from_list(self, item):
        if item.text() in self.cloud_free_folders_paths:
            self.cloud_free_folders_paths.remove(item.text())
            self.update_selected_cloud_free_files_list()

    def load_and_display_images(self):
        if not self.cloudy_folders_paths or not self.cloud_free_folders_paths:
            print("Please select both sets of folders.")
            return

        for cloudy_folder_path, cloud_free_folder_path in zip(self.cloudy_folders_paths, self.cloud_free_folders_paths):
            # Extract files from the cloudy folder
            cloudy_id = os.path.splitext(os.path.basename(cloudy_folder_path))[0]
            tar = tarfile.open(cloudy_folder_path)
            tar.extractall(cloudy_id)
            tar.close()

            # Extract files from the cloud-free folder
            cloud_free_id = os.path.splitext(os.path.basename(cloud_free_folder_path))[0]
            tar = tarfile.open(cloud_free_folder_path)
            tar.extractall(cloud_free_id)
            tar.close()

            # Count the number of files in the extracted folders
            num_cloudy_files = len(os.listdir(cloudy_id))
            num_cloud_free_files = len(os.listdir(cloud_free_id))

            print("Cloudy Folder ID:", cloudy_id)
            print(f"Number of Cloudy Files: {num_cloudy_files}")
            print("Cloud-Free Folder ID:", cloud_free_id)
            print(f"Number of Cloud-Free Files: {num_cloud_free_files}")
            print("Images extracted successfully.")
            
            output_folder = 'output'
            # Load and display the images
            cloudy_image = self.get_RGB(cloudy_id)
            print("Dowloading Cloud Image")
            tiff.imsave(f'{output_folder}/{cloudy_id}_cloud.tif', cloudy_image)
            print("Cloud Image Downloaded")
            
            
            non_cloud_image = self.get_RGB(cloud_free_id)
            print("Downloading Non Cloud Image")
            tiff.imsave(f'{output_folder}/{cloud_free_id}_non_cloud.tif', non_cloud_image)
            print("Non Cloud Image Downloaded")

            if cloudy_image is not None and non_cloud_image is not None:
                print("Images loaded successfully.")
            else:
                print("Failed to load images.")
                return

            print("Cloud Remove is processing")

            # Get cloud mask
            cloud_mask = self.cloud_pred(cloudy_id)

            # Remove clouds
            rm_cloud = cloudy_image * cloud_mask[:, :, np.newaxis]

            # Clip the pixel values in rm_cloud to the range [0, 255]
            rm_cloud = np.clip(rm_cloud, 0, 255).astype(np.uint8)
            print("Remove Cloud Successfully!")
            
            print("Downloading Removed Cloud")
            tiff.imsave(f'{output_folder}/{cloudy_id}_removed_cloud.tif', rm_cloud)
            print("Removed Cloud Downloaded")
            
            print("Cloud Mask is processing")
            geo_cloudy = rio.open(f'{cloudy_id}/{cloudy_id}_SR_B4.TIF')
            geo_clear = rio.open(f'{cloud_free_id}/{cloud_free_id}_SR_B4.TIF')

            QA = tiff.imread(f'{cloudy_id}/{cloudy_id}_QA_PIXEL.TIF')
            QA = np.array(QA)

            cloud_mask = np.vectorize(self.get_mask)(QA, type='cloud')
            shadow_mask = np.vectorize(self.get_mask)(QA, type='shadow')
            dilated_cloud_mask = np.vectorize(self.get_mask)(QA, type='dilated_cloud')
            cirrus_mask = np.vectorize(self.get_mask)(QA, type='cirrus')

            mask = cloud_mask * shadow_mask * dilated_cloud_mask * cirrus_mask
            print("Cloud Mask is Successfully")

            print("Remove Mask is processing")
            rm_mask = cloudy_image * mask[:, :, np.newaxis]
            print("Remove Mask is Successfully")
            
            print("Getting UTM Coordinates")
            # get UTM coordinates from cloudy image
            utmx, utmy = geo_cloudy.xy(cloudy_image.shape[0], cloudy_image.shape[1])

            # get pixels from clear image
            y_, x_ = geo_clear.index(utmx, utmy)

            # get pixel coordinates of non-cloudy image
            y, x = geo_clear.read(1).shape
            clear_ul = (0, 0)  # upper left
            clear_lr = (y, x)  # lower right

            # get pixel coordinates of cloudy image corners
            cloudy_bounds = geo_cloudy.bounds
            new_clear_ul = geo_clear.index(cloudy_bounds.left, cloudy_bounds.top)
            new_clear_lr = geo_clear.index(cloudy_bounds.right, cloudy_bounds.bottom)

            # calculate pixel adjustment
            top_adj = clear_ul[0] - new_clear_ul[0]
            bottom_adj = new_clear_lr[0] - clear_lr[0]

            left_adj = clear_ul[1] - new_clear_ul[1]
            right_adj = new_clear_lr[1] - clear_lr[1]
            
            print("UTM Coordinates is Successfully")
            
            print("Inpainting is processing")
            # Adjust clear RGB image
            clear_RGB_adj = self.adjust_rgb(non_cloud_image, top_adj, bottom_adj, left_adj, right_adj)
            fill_mask = clear_RGB_adj * (1 - mask[:, :, np.newaxis])
            inpaint = rm_mask + fill_mask
            inpaint = np.clip(inpaint, 0, 255).astype(np.uint8)
            print("Inpainting Successfully")
            
            # Replace 'output_folder' with your desired output folder path

            print("Downloading Image")
            # Save inpaint as a TIF file
            tiff.imsave(f'{output_folder}/{cloudy_id}_inpaint.tif', inpaint)
            print("Download Successfully")
            
            self.display_images(cloudy_image, non_cloud_image, rm_cloud, inpaint)
        
    def get_RGB(self, ID):
        # Load Blue (B2), Green (B3), and Red (B4) bands
        B2 = tiff.imread(f'{ID}/{ID}_SR_B2.TIF')
        B3 = tiff.imread(f'{ID}/{ID}_SR_B3.TIF')
        B4 = tiff.imread(f'{ID}/{ID}_SR_B4.TIF')

        # Stack the bands
        RGB = np.dstack((B4, B3, B2))
        
        # Clip to enhance contrast
        
        RGB = np.clip(RGB*0.0000275-0.2, 0, 1)
        
        RGB = np.clip(RGB,0,0.3)/0.3
        
        min_val = RGB.min()
        max_val = RGB.max()
        RGB = ((RGB - min_val) / (max_val - min_val) * 255).astype(np.uint8)

        return RGB
    
    def cloud_pred(self, ID):
        """Cloud prediction model"""
        B1 = tiff.imread(f'{ID}/{ID}_SR_B1.TIF')
        B4 = tiff.imread(f'{ID}/{ID}_SR_B4.TIF')
        B7 = tiff.imread(f'{ID}/{ID}_SR_B7.TIF')

        pred = 2.16246741593412 -0.796409165054949*B4 + \
        0.971776520302587*np.sqrt(abs(0.028702220187686*B7*B1 + \
        0.971297779812314*np.sin(B1))) + \
        0.0235599298084993*np.floor(0.995223926146334*np.sqrt(abs(0.028702220187686*B7*B1 + 0.971297779812314*np.sin(B1)))+ \
        0.00477607385366598*abs(0.028702220187686*B7*B1 + \
        0.971297779812314*np.sin(B1))) - 0.180030905136552*np.cos(B4) + \
        0.0046635498889134*abs(0.028702220187686*B7*B1 + 0.971297779812314*np.sin(B1))

        cloud = np.where(pred > B7,0,1)

        return cloud
    
    def get_mask(self, val,type='cloud'):

        """Get mask for a specific cover type"""

        # convert to binary
        bin_ = '{0:016b}'.format(val)

        # reverse string
        str_bin = str(bin_)[::-1]

        # get bit for cover type
        bits = {'cloud':3,'shadow':4,'dilated_cloud':1,'cirrus':2}
        bit = str_bin[bits[type]]

        if bit == '1':
            return 0 # cover
        else:
            return 1 # no cover
        
    def adjust_rgb(self, rgb,top_adj, bottom_adj, left_adj, right_adj):

        adj_rgb = rgb.copy()

        #Adding black pixels
        if top_adj > 0:
            add_top = np.zeros((top_adj,rgb.shape[1],3))
            adj_rgb = np.vstack((add_top,adj_rgb))
        if bottom_adj > 0:
            add_bottom = np.zeros((bottom_adj,rgb.shape[1],3))
            adj_rgb = np.vstack((adj_rgb,add_bottom))
        if left_adj > 0:
            add_left = np.zeros((rgb.shape[0],left_adj,3))
            adj_rgb = np.hstack((add_left,adj_rgb))
        if right_adj > 0:
            add_right = np.zeros((rgb.shape[0],right_adj,3))
            adj_rgb = np.hstack((adj_rgb,add_right))

        #Removing pixels
        if top_adj < 0:
            adj_rgb = adj_rgb[-top_adj:,:,:]
        if bottom_adj < 0:
            adj_rgb = adj_rgb[:bottom_adj,:,:]
        if left_adj < 0:
            adj_rgb = adj_rgb[:,-left_adj:,:]
        if right_adj < 0:
            adj_rgb = adj_rgb[:,:right_adj,:]

        return adj_rgb

    def display_images(self, cloudy_image, non_cloud_image, rm_cloud_image, inpaint_image):
        self.scene.clear()

        # Create QGraphicsPixmapItems for the images
        cloudy_item = QGraphicsPixmapItem(QPixmap.fromImage(self.ndarray_to_qimage(cloudy_image)))
        non_cloud_item = QGraphicsPixmapItem(QPixmap.fromImage(self.ndarray_to_qimage(non_cloud_image)))
        rm_cloud_item = QGraphicsPixmapItem(QPixmap.fromImage(self.ndarray_to_qimage(rm_cloud_image)))
        inpaint_item = QGraphicsPixmapItem(QPixmap.fromImage(self.ndarray_to_qimage(inpaint_image)))

        # Calculate the aspect ratio of the images
        aspect_ratio = float(cloudy_image.shape[0]) / cloudy_image.shape[1]

        # Calculate the maximum width and height for all images
        max_width = max(cloudy_image.shape[1], non_cloud_image.shape[1], rm_cloud_image.shape[1], inpaint_image.shape[1])
        max_height = int(max_width * aspect_ratio)

        # Position the items next to each other without any space in between
        x = 0
        y = 0

        for item in [cloudy_item, non_cloud_item, rm_cloud_item, inpaint_item]:
            item.setPos(x, y)
            x += max_width  # Position the next item to the right

        # Add the items to the scene
        self.scene.addItem(cloudy_item)
        self.scene.addItem(non_cloud_item)
        self.scene.addItem(rm_cloud_item)
        self.scene.addItem(inpaint_item)

        # Set the scene rect to fit all four images
        self.graphics_view.setSceneRect(0, 0, max_width * 4, max_height)

        # Update the graphics view to display the images
        self.graphics_view.update()
    
    def ndarray_to_qimage(self, array):
        height, width, channel = array.shape
        bytes_per_line = 3 * width
        # Ensure the data type is 8-bit (0-255) if it's not already
        if array.dtype != np.uint8:
            array = (array * 255).astype(np.uint8)
        qimage = QImage(array.data.tobytes(), width, height, bytes_per_line, QImage.Format_RGB888)
        return qimage


class CustomGraphicsView(QGraphicsView):
    def __init__(self, parent=None):
        super(CustomGraphicsView, self).__init__(parent)
        self.setRenderHints(QPainter.Antialiasing | QPainter.SmoothPixmapTransform)
        self.setInteractive(True)
        self.setRenderHint(QPainter.Antialiasing)

    def wheelEvent(self, event):
        # Enable zooming with the mouse wheel
        zoom_in_factor = 1.25
        zoom_out_factor = 1 / zoom_in_factor

        # Zoom in or out based on the wheel movement
        if event.angleDelta().y() > 0:
            self.scale(zoom_in_factor, zoom_in_factor)
        else:
            self.scale(zoom_out_factor, zoom_out_factor)

    def mousePressEvent(self, event):
        # Pan the view using the mouse click and drag
        if event.button() == Qt.MiddleButton:
            self.setDragMode(QGraphicsView.ScrollHandDrag)
            self.setInteractive(True)

    def mouseReleaseEvent(self, event):
        if event.button() == Qt.MiddleButton:
            self.setDragMode(QGraphicsView.NoDrag)
            self.setInteractive(True)

def main():
    app = QApplication(sys.argv)
    window = ImageProcessingApp()
    window.show()
    sys.exit(app.exec_())

if __name__ == '__main__':
    main()
```
